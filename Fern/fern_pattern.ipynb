```python
import numpy as np
import random as ra
import matplotlib.pyplot as plt
# import numba
```

```python
# @numba.njit(cache = True)
def fast_transform(n = 1000):

    r0 = [0,0] # initial array
    resultt = []

    for i in range(n):
        # print(f"Checking {i}", end="\r")
        z=ra.randint(0,3)
        if z==0:
            r0 = [0, 0.16*r0[1]]
        elif z==1:
            r0 =  [0.85*r0[0] + 0.04*r0[1],-0.04*r0[0]+0.85*r0[1]+1.6]
        elif z==2:
            r0 = [0.2*r0[0]-0.26*r0[1],0.23*r0[0]+0.22*r0[1]+1.6]
        elif z==3:
            r0 = [-0.15*r0[0]+0.28*r0[1], 0.26*r0[0]+0.24*r0[1]+0.44]
        resultt.append(r0)
    return resultt
```

```python
result = fast_transform(n = 50000)
result = np.array(result)
```

```python
plt.scatter(result[:, 0], result[:, 1], s=[0.5], c = 'g')
```
